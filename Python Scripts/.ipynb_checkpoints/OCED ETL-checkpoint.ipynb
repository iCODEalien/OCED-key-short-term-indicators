```python
# Dataframe Object
oced_df = pd.read_csv(r'C:\Users\Chaitra.b.c\Desktop\Data Anaytics\Dataset\6052.csv',
                     parse_dates = [0]);
oced_df.info();
```

```python
# For percentage calculation
#len(oced_df)

not_null_values = len(oced_df) - oced_df.isna().sum()
prc_not_null_values = (not_null_values/len(oced_df))*100
not_null_series = pd.Series(prc_not_null_values)
#not_null_series

null_series = oced_df.isna().sum()
null_prc = (null_series.values/len(oced_df))*100
```

```python
# missing values count
#oced_df.isna().sum().index


plt.figure(figsize = (10,20))
plt.barh(y = oced_df.isna().sum().index, width = oced_df.isna().sum().values, height = 0.3,label = 'Null values')
plt.barh(y = oced_df.isna().sum().index, width = 36775 - oced_df.isna().sum().values, height = 0.7 ,label = 'Non Null values') # non null values


for i,j in enumerate(zip(not_null_values.values, not_null_series.values,null_series.values, null_prc)):
    #print(j[2]/2,j[3]/2)
    plt.text(y = i, x = j[0]/2, s = str(np.round(j[1],2)) + '%')
    plt.text(y = i, x = j[2]/2, s = str(np.round(j[3],2)) + '%')
    
plt.legend()
plt.title('Stacked Bar chart for missing values')
plt.grid(color = 'black', axis = 'x', alpha = 0.1)
plt.show()
```

```python
# 
df = pd.DataFrame(pd.Series(oced_df.columns), columns = ['Columns'])
df['not_null_values'] = pd.Series(oced_df.isna().sum().values)
df['null_values'] = pd.Series(len(oced_df) - oced_df.isna().sum().values)

df
```

```python
plt.figure(figsize = (10,20))
plt.barh(width = df['not_null_values'], y = df['Columns'].index,  height = 0.35, label = ' Not nUll values')
plt.barh(width = df['null_values'], y = df['Columns'].index + 0.35, height = 0.2, label = 'Null values')
plt.yticks(ticks = range(0,len(oced_df.columns)), labels = list(oced_df.columns))
plt.legend()
plt.show()
```
